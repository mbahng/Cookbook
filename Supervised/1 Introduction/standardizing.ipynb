```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler, QuantileTransformer

wine_data = pd.read_csv("https://raw.githubusercontent.com/derekbanas/tensorflow/main/winequalityN.csv") 
wine_data = wine_data.dropna() 
del wine_data["type"]
X = wine_data.to_numpy()[::10, :2]
print(X.shape)
```

### Standard Scaler

```python
scaler = StandardScaler() 
scaled_data = scaler.fit_transform(X)
plt.scatter(*zip(*scaled_data), c="r", s=3)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
# plt.savefig("Standard_scaler1.png")
plt.show()
```

```python
scaler = MinMaxScaler() 
scaled_data = scaler.fit_transform(X)
plt.scatter(*zip(*scaled_data), c="r", s=3)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.savefig("MinMax_Scaler.png")
```

```python
scaler = MaxAbsScaler() 
scaled_data = scaler.fit_transform(X)
plt.scatter(*zip(*scaled_data), c="r", s=3)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.savefig("MaxAbs_Scaler.png")
```

```python
scaler = RobustScaler() 
scaled_data = scaler.fit_transform(X)
plt.scatter(*zip(*scaled_data), c="r", s=3)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.savefig("Robust_Scaler.png")
```

```python
scaler = QuantileTransformer(output_distribution="normal") 
scaled_data = scaler.fit_transform(X)
plt.scatter(*zip(*scaled_data), c="r", s=3)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.savefig("Quantile_Scaler.png")
```
