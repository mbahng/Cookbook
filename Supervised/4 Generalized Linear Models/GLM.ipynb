```python
import numpy as np 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
import scipy.stats as stats 
```

```python
X = np.sort(np.random.uniform(0, 2, 8) )
Y = np.array([0, 0, 0, 0, 1, 1,1,1])

model = LinearRegression().fit(X.reshape(-1, 1), Y.reshape(-1, 1))

plt.scatter(X, Y, c="b", s=4, label="Dataset")
rang = np.linspace(-1, 3, 100)
plt.plot(rang, model.predict(rang.reshape(-1, 1)), c="r", linewidth=2, label="OLS Model")
plt.savefig("GLM_bernoulli.png")
plt.show() 
```

```python
from scipy.stats import norm 

def sigmoid(x:np.array): 
    return 1/(1 + np.exp(-x)) 

def norm_cdf(x:np.array): 
    return norm.cdf(x) 

x = np.linspace(-5, 5, 100) 
plt.plot(sigmoid(x), x, label="Logit")
plt.plot(norm_cdf(x), x, label="Probit")
plt.xlabel("mu(x)")
plt.ylabel("g(mu(x))")
plt.legend()
plt.savefig("probit_logit.png")
plt.show()
```
